In the test folder, there are just 7 images which are very less. So,what I have done is , copied 20 random images from each subdirectory in the train folder.

Firstly, I generated a list of 20 random integers,which would be the indexes of images to be copied.

```
import random
randomImageIndex = random.sample(range(1, 100), 20)
```

**Copying images from train folder to test folder**

**Shutil Module -->**
The shutil module provides functions for copying files, as well as entire folders. For copying multiple files at once, we have to have a list of all files you want to copy and loop over them to copy them.

**Calling shutil.copy(source, destination)** will copy the file at the path source to the folder at the path destination. (Both source and destination are strings.) If destination is a filename, it will be used as the new name of the copied file.

```
import shutil
train_dir=os.listdir("mobile_gallery_image_classification/train")
for fold in train_dir:
  image_fold=os.listdir("mobile_gallery_image_classification/train/" +fold)
  for idx in randomImageIndex:
    img="mobile_gallery_image_classification/train/"+fold+"/"+image_fold[idx]
    shutil.copy( img, "mobile_gallery_image_classification/test")
```

After modifying the dataset according to our need, now the task of uploading the dataset to my kaggle id comes.

One thing to be done before creating a new dataset and uploading on kaggle id

**Initialize metadata file for dataset creation**

This can be done by running  **kaggle datasets init [-h] [-p FOLDER]** command

where  **-h, --help**           show this help message and exit

**-p FOLDER, --path FOLDER**  Folder for upload, containing data files and a special dataset-metadata.json file. Defaults to current working directory

```
!kaggle datasets init -p "mobile_gallery_image_classification"
```

**Create a new dataset**

I have used ,

kaggle datasets create [-h] [-p FOLDER] [-u] [-q] [-t] [-r {skip,zip,tar}]

**optional arguements :**

**-u, --public -->** Create publicly (default is private)

**-r {skip,zip,tar}, --dir-mode {skip,zip,tar} -->** What to do with directories: "skip" - ignore; "zip" - compressed upload; "tar" - uncompressed upload

**-p FOLDER, --path FOLDER -->** Folder for upload, containing data files and a special dataset-metadata.json file. Defaults to current working directory

**-t, --keep-tabular -->** Do not convert tabular files to CSV (default is to convert)

**-q, --quiet -->** Suppress printing information about the upload/download progress
  
**-h, --help -->**   show this help message and exit

```
!kaggle datasets create -p "mobile_gallery_image_classification" -u -r zip  --dir-mode zip
```
